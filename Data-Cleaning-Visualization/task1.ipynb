```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("C:/Users/USER/Downloads/job_salary_prediction_dataset.csv")
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
df.columns = df.columns.str.strip()
df.drop_duplicates(inplace=True)
df.fillna(method='ffill', inplace=True)
df['experience_years'] = df['experience_years'].astype(int)
df['salary'] = df['salary'].astype(float)
```

1. Experience vs Salary

```python
sns.scatterplot(x='experience_years', y='salary', data=df)
plt.title("Experience vs Salary")
plt.show()
```

Salary by Job Role

```python
plt.figure(figsize=(10,5))
sns.barplot(x='job_title', y='salary', data=df)
plt.xticks(rotation=45)
plt.title("Salary by Job Role")
plt.show()
```

Remote Work Impact

```python
sns.boxplot(x='remote_work', y='salary', data=df)
plt.title("Remote Work vs Salary")
plt.show()
```

Education vs Salary

```python
sns.barplot(x='education_level', y='salary', data=df)
plt.title("Education vs Salary")
plt.show()
```

Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
